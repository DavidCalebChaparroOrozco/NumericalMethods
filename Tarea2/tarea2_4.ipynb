```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
# Definir la funci√≥n de Runge
def runge(x):
    return 1 / (1 + 25 * x ** 2)
```

```python
# 1.- Graficar la funci√≥n de Runge en el intervalo [-1,1]
x = np.linspace(-1, 1, 100)
y = runge(x)
plt.plot(x, y)
plt.title("Funci√≥n de Runge")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 2.- Generar y graficar el polinomio de interpolaci√≥n de Lagrange
xi = np.array([-1, -0.5, 0, 0.5, 1])
yi = runge(xi)
poly = lagrange(xi, yi)
x = np.linspace(-1, 1, 100)
y = poly(x)
plt.plot(x, y, label="Polinomio de Lagrange")
plt.plot(xi, yi, "o", label="Datos equiespaciados")
plt.plot(x, runge(x), label="Funci√≥n de Runge")
plt.title("Polinomio de Lagrange para la Funci√≥n de Runge")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# 3.- Estimar f(0.8) con polinomios de interpolaci√≥n de grado 2 y 3
poly2 = lagrange(xi[1:4], yi[1:4])
poly3 = lagrange(xi, yi)
print("El valor estimado de f(0.8) con polinomio de grado 2 es:", poly2(0.8))
print("El valor estimado de f(0.8) con polinomio de grado 3 es:", poly3(0.8))
```
